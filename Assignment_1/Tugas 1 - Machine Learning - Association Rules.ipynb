```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Mengimport data set
data_retail= pd.read_excel('https://drive.google.com/uc?export=download&id=11WL3jd8GAjR7RxQE01eWcE6JY_JsjLM0')

# Menampilkan tampilan awal dataset
data_retail
```

```python
# Memangkas beberapa kolom yang tidak perlu
data_retail= data_retail[['Invoice','Description', 'Quantity', 'Country']]

# Memfokuskan pada Country United Kingdom
data_retail= data_retail[data_retail['Country']=='United Kingdom']

data_retail
```

```python
# Mengecek jika ada nilai NaN pada data_retail
data_retail.isna().sum()
```

```python
# Menghapus baris dengan description bernilai NaN
data_retail.dropna()

# Membersihkan spasi yang tidak perlu pada data_retail
data_retail['Description']= data_retail['Description'].str.strip()

# Fungsi groupby dan unstack() untuk mengubah dimensi data_retail menjadi invoice x description
# Juga membuat invoice menjadi index dengan fungsi reset_index() lalu set_index()
# Fungsi fillna(0) mengganti nilai NaN menjadi 0
 
data_retail_sorted= (data_retail.groupby(['Invoice','Description'])['Quantity']
                .sum().unstack().reset_index().fillna(0).set_index('Invoice'))

data_retail_sorted
```

```python
# Membuat fungsi encode

def quantity_encode(x):
    if(x<=0):
        return 0
    if(x>=1):
        return 1
```

```python
# Mengaplikasikan fungsi quantity_encode pada data_retail
# Fungsinya mengubah data menjadi 0 dan 1, 0 jika tidak dibeli, 1 jika dibeli berapapun quintitynya

data_retail_sorted= data_retail_sorted.applymap(quantity_encode)
data_retail_sorted
```

```python
# Membangun model apriori
# Dengan minimum support 1.75%, rendah karena column deskripsi berjumlah 4000
apriori_data_retail= apriori(data_retail_sorted, min_support=0.0175, use_colnames=True)

# Membuat association rules
arules= association_rules(apriori_data_retail, metric='lift', min_threshold=1)

# Menampilkan association rules dengan confidence dan lift paling tinggi
arules.sort_values(['confidence','lift'], ascending=[False,False])
```
